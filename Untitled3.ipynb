```python
import pickle
import pandas as pd
from os import path
%matplotlib inline
```

```python
file_name = 'df_results.pickle'
file_path = path.join('df_results', file_name)
```

```python
df_results = pickle.load(open(file_path, 'rb'))
```

```python
df_results.columns
```

```python
pd.set_option('display.max_columns', 100)
```

```python
df_results.head(3)
```

```python
filt = df_results['status_type'] == 'PASSED'
```

```python
counts = df_results.loc[filt].groupby('operator')['status_type'].count()
```

```python
counts.index = counts.index.str.lower()
```

```python
counts
```

```python
counts.sort_index().plot.bar(xlabel='System', ylabel='Passed', figsize=(10,10))
```
