```python
import pandas as pd
import os
```

```python
file_path = os.path.join('Result Files', '6171-R&D-00000077_CEQ0134.csv')
```

```python
df = pd.read_csv(file_path, header=2)
```

```python
df
```

```python
filt = df['Passed'] == True
```

```python
df_passed = df.loc[filt]
```

```python
df_passed_unique = df_passed.groupby('Serial').last().reset_index()
```

```python
with open('df_passed_unique_RND_build_77.csv', 'w') as file:
    file.write(df_passed_unique.to_csv(index=False, line_terminator="\n"))
```
